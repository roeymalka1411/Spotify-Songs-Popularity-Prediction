# Initial Data Cleaning

## Deleting Null Indecies:


```python
# show me the indecies of the rows with null values:
null_indices = df[df.isnull().any(axis=1)].index

# 5 rows with null values:
display(df[df.index.isin(null_indices)])

# deleting those rows (don't have name, artist, album and they have 0 popularity, so they are not useful for our analysis)
df = df[~df.index.isin(null_indices)]
```

## Handling duplicates
* Duplicates will be determined by "str type" columns.
* Before removing a duplicate, its "numric cols" will be averaged.

```python
dup_cols = [
    'track_name', 
    'track_artist', 
    'track_album_name', 
    'playlist_name', 
    'playlist_genre', 
    'playlist_subgenre'
]

df.sort_values(by = 'track_popularity', ascending=False, inplace=True)

rows = len(df)
df.drop_duplicates(subset=dup_cols, keep='first', inplace=True)

print(f"{rows - len(df)} duplicate rows dropped. {len(df)} rows remaining.")
```

## Removing useless columns
* Irrelevant cols: ['track_id', 'track_name', 'track_album_id', 'playlist_id']

```python
cols_to_drop = ['track_id', 'track_name', 'track_album_id', 'playlist_id']
df.drop(columns=cols_to_drop, inplace=True)
```

## Basic Fearture Engineering
* **track age**: Extracted and calculated from the original **track_album_release_date** to represent the temporal age of the song, transforming a static date string into a continuous numerical variable.

* **duration_sec**: Converting **duration_ms** into **duration_sec** by splitting it in 1000.

* **One-Hot Encoding (OHE)**: Applied strictly to the low-cardinality playlist_genre feature (dropping the first category to avoid the dummy variable trap). This allows linear models to assign direct, independent weights to the primary genres.
 
* **Cyclical Encoding**: Applied sine and cosine transformations to the key column (pitch classes 0-11). This maps the keys onto a 2D continuous circle, preserving their inherent musical circularity (where key 11 is adjacent to key 0) while avoiding target leakage and the dimensionality bloat of OHE.

### track_age and duration_sec

```python
# calculate the age of the track in years based on the release date:
df['track_release_date'] = pd.to_datetime(df['track_album_release_date'], errors='coerce')
df['track_age'] = (pd.to_datetime('today') - df['track_release_date']).dt.days / 365

# Converting 'duration_ms' into 'duration_sec':
df['duration_sec'] = df['duration_ms'] / 1000

df.drop(columns = ['track_release_date', 'track_album_release_date'], inplace = True)
df.drop(columns = ['duration_ms'], inplace = True)
```

### OHE ('playlist_genre' column)

```python
# One-Hot Encoding for playlist_genre ---
# Apply OHE with drop_first=True to avoid multicollinearity (Dummy Variable Trap)
df = pd.get_dummies(df, columns=['playlist_genre'], drop_first=True)

genre_cols = [col for col in df.columns if "playlist_genre_" in col]
df[genre_cols] = df[genre_cols].astype('int')
```

### Cyclical Encoding

```python
# Define the maximum value for the cycle (12 pitch classes, 0 to 11)
max_key = 12

# Create the Sine transformation:
df['key_sin'] = np.sin(2 * np.pi * df['key'] / max_key)
# Create the Cosine transformation:
df['key_cos'] = np.cos(2 * np.pi * df['key'] / max_key)

df.drop(columns = ['key'], inplace = True)
```
